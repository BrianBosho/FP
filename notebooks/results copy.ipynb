```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python

# Read the CSV file into a DataFrame
df = pd.read_csv('results.csv')

# Extract the client_id, training_losses, and training_accuracies columns
client_ids = df['client_id']
training_losses = df['training_losses'].apply(ast.literal_eval)
training_accuracies = df['training_accuracies'].apply(ast.literal_eval)
```

```python



# Create a dictionary to hold DataFrames for each client
losses_dict = {}
accuracies_dict = {}

# Populate the dictionaries with DataFrames for each client
for client_id, losses, accuracies in zip(client_ids, training_losses, training_accuracies):
    losses_dict[f'client_{client_id}'] = losses
    accuracies_dict[f'client_{client_id}'] = accuracies

# Convert the dictionaries to DataFrames
df_losses = pd.DataFrame(losses_dict)
df_accuracies = pd.DataFrame(accuracies_dict)

# Print the DataFrames
# print("Training Losses DataFrame:")
# print(df_losses.head())
# print("\nTraining Accuracies DataFrame:")
# print(df_accuracies.head())
```

```python
# Plot training losses
plt.figure(figsize=(12, 6))
for column in df_losses.columns:
    plt.plot(df_losses.index, df_losses[column], label=column)
plt.title('Training Losses')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot training accuracies
plt.figure(figsize=(12, 6))
for column in df_accuracies.columns:
    plt.plot(df_accuracies.index, df_accuracies[column], label=column)
plt.title('Training Accuracies')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
# add grid
plt.grid()
# add grid with steps of 5 on the horizontal axis

plt.show()
```
