## 1. Load Data

```python
import os
# notebook lives in notebooks/; resolve paths relative to repo root
REPO_ROOT = '/home/bosho/FP'
os.chdir(REPO_ROOT)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

plt.rcParams.update({'figure.dpi': 120, 'font.size': 10, 'axes.spines.top': False, 'axes.spines.right': False})
pd.set_option('display.max_colwidth', None)

BASE  = 'experiments/output'
R1_CSV  = f'{BASE}/R1_consolidated.csv'
R1B_CSV = f'{BASE}/R1b_consolidated.csv'

r1  = pd.read_csv(R1_CSV)
r1b = pd.read_csv(R1B_CSV)

# Convert fractional accuracy → percentage
r1 ['mean_pct'] = r1 ['mean_acc'] * 100
r1b['mean_pct'] = r1b['mean_acc'] * 100

print(f'R1  — {len(r1)} rows  ({r1["dataset"].nunique()} datasets, {r1["model"].nunique()} models)')
print(f'R1b — {len(r1b)} rows  ({r1b["dataset"].nunique()} datasets, {r1b["model"].nunique()} models)')
```

## 2. Data Availability — What We Have vs What's Missing

Full design matrix: 3 datasets × 3 βs × 4 propagation methods × (PE variants). 

```python
# ── helper ─────────────────────────────────────────────────────────────────────
BETAS = [10000, 10, 1]
BETA_LABELS = {10000: 'β=10000 (IID)', 10: 'β=10', 1: 'β=1 (high non-IID)'}
PROPS_R1  = ['zero_hop', 'full', 'adjacency', 'diffusion']   # R1 no-PE
PROPS_R1B = ['zero_hop', 'full', 'adjacency', 'diffusion']   # R1b no-PE + PE variant
```

```python
# ── R1 (GCN, hop-1, no PE) — completeness check ─────────────────────────────
print('═══ R1 — GCN hop-1 (no PE) ═══')
datasets_r1 = ['Cora', 'Citeseer', 'Pubmed']
props_r1    = ['zero_hop', 'full', 'adjacency', 'diffusion']

for ds in datasets_r1:
    sub = r1[r1['dataset'] == ds]
    print(f'\n{ds}:')
    for prop in props_r1:
        row_data = {}
        for beta in BETAS:
            match = sub[(sub['propagation'] == prop) & (sub['beta'] == beta)]
            if len(match) == 0:
                row_data[BETA_LABELS[beta]] = '—'
            else:
                n = match.iloc[0]['n_reps']
                acc = match.iloc[0]['mean_pct']
                std = match.iloc[0]['std_acc'] * 100
                flag = f' ⚠{n}' if n < 10 else ''
                row_data[BETA_LABELS[beta]] = f'{acc:.1f}±{std:.1f}{flag}'
        print(f'  {prop:12s}: {row_data}')
```

```python
# ── R1b (GAT, hop-2, no-PE and PE) — completeness check ─────────────────────
print('═══ R1b — GAT hop-2 (with/without PE) ═══')
datasets_r1b = ['Cora', 'Citeseer', 'Pubmed']
props_r1b    = ['zero_hop', 'full', 'adjacency', 'diffusion']

for ds in datasets_r1b:
    sub = r1b[r1b['dataset'] == ds]
    print(f'\n{ds}:')
    for prop in props_r1b:
        for pe in [False, True]:
            pe_str = 'PE' if pe else 'no PE'
            row_data = {}
            for beta in BETAS:
                match = sub[(sub['propagation'] == prop) & (sub['beta'] == beta) & (sub['use_pe'] == pe)]
                if len(match) == 0:
                    row_data[BETA_LABELS[beta]] = '—'
                else:
                    n   = match.iloc[0]['n_reps']
                    acc = match.iloc[0]['mean_pct']
                    std = match.iloc[0]['std_acc'] * 100
                    flag = f' ⚠{n}' if n < 10 else ''
                    row_data[BETA_LABELS[beta]] = f'{acc:.1f}±{std:.1f}{flag}'
            print(f'  {prop:12s} ({pe_str}): {row_data}')
```

## 3. R1 — GCN hop-1 (no PE)

```python
# ── R1: Cora — bar chart by method, grouped by β ─────────────────────────────
ds = 'Cora'
sub = r1[r1['dataset'] == ds].copy()

BETA_MAP = {10000: 'b10000', 10: 'b10', 1: 'b1'}
BETA_LABELS_LIST = ['b10000', 'b10', 'b1']
BETA_VALS = {'b10000': 10000, 'b10': 10, 'b1': 1}

METHOD_COLORS_R1 = {
    'zero_hop':  '#aec7e8',
    'full':      '#1f77b4',
    'adjacency': '#98df8a',
    'diffusion': '#ffbb78',
}
METHOD_LABELS = {
    'zero_hop':  'Zero-hop',
    'full':      'Full-graph',
    'adjacency': 'Adj (no PE)',
    'diffusion': 'Diff (no PE)',
}

fig, ax = plt.subplots(figsize=(10, 5))
betas_str = BETA_LABELS_LIST
x = np.arange(len(betas_str))
width = 0.18
n = len(METHOD_COLORS_R1)
offsets = np.linspace(-(n-1)/2, (n-1)/2, n) * width

for i, (prop, color) in enumerate(METHOD_COLORS_R1.items()):
    means, stds = [], []
    for beta_str in betas_str:
        beta_val = BETA_VALS[beta_str]
        row = sub[(sub['propagation']==prop) & (sub['beta']==beta_val)]
        if len(row):
            means.append(row.iloc[0]['mean_pct'])
            stds.append(row.iloc[0]['std_acc'] * 100)
        else:
            means.append(np.nan); stds.append(0)
    bars = ax.bar(x + offsets[i], means, width, label=METHOD_LABELS[prop],
                 color=color, alpha=0.85, yerr=stds, capsize=3, error_kw={'elinewidth':1})
    for j, m in enumerate(means):
        if np.isnan(m): bars[j].set_alpha(0)

ax.set_xlabel('Dirichlet beta  (10000 = IID  ->  1 = high non-IID)')
ax.set_ylabel('Test Accuracy (%)')
ax.set_title(f'{ds} - R1 GCN hop-1 (SGD lr=0.5)', fontsize=12, fontweight='bold')
ax.set_xticks(x); ax.set_xticklabels(betas_str)
ax.set_ylim(50, 90)
ax.grid(axis='y', alpha=0.3)
ax.legend(fontsize=9, bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## 4. R1b — GAT hop-2 (no PE + PE variants)

```python
# ── R1b: Cora — grouped by method+PE, bars = beta ───────────────────────────
ds = 'Cora'
sub = r1b[r1b['dataset'] == ds].copy()

fig, ax = plt.subplots(figsize=(14, 5))

# Each (prop, pe) pair gets its own x-slot; groups of 3 betas side by side
method_slots = [
    ('zero_hop',  False),
    ('full',      False),
    ('adjacency', False),
    ('adjacency', True),
    ('diffusion', False),
    ('diffusion', True),
]
slot_labels = [
    'Zero-hop', 'Full-graph',
    'Adj (no PE)', 'Adj (PE)',
    'Diff (no PE)', 'Diff (PE)',
]
slot_colors = [
    '#aec7e8', '#1f77b4',
    '#98df8a', '#2ca02c',
    '#ffbb78', '#ff7f0e',
]

n_slots = len(method_slots)
x = np.arange(n_slots)
width = 0.22
beta_colors = {'b10000': '#4C72B0', 'b10': '#DD8452', 'b1': '#55A868'}
offsets_map = {-width: 'b10000', 0: 'b10', width: 'b1'}

for off, beta_str in offsets_map.items():
    beta_val = BETA_VALS[beta_str]
    means, stds = [], []
    for prop, pe in method_slots:
        row = sub[(sub['propagation']==prop) & (sub['beta']==beta_val) & (sub['use_pe']==pe)]
        if len(row):
            means.append(row.iloc[0]['mean_pct'])
            stds.append(row.iloc[0]['std_acc']*100)
        else:
            means.append(np.nan); stds.append(0)
    bars = ax.bar(x + off, means, width, label=beta_str,
                  color=beta_colors[beta_str], alpha=0.85,
                  yerr=stds, capsize=3, error_kw={'elinewidth':1})
    for j, m in enumerate(means):
        if np.isnan(m): bars[j].set_alpha(0)

ax.set_xlabel('Method')
ax.set_ylabel('Test Accuracy (%)')
ax.set_title(f'{ds} - R1b GAT hop-2 (SGD lr=0.5)', fontsize=12, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(slot_labels, rotation=20, ha='right')
ax.set_ylim(50, 90)
ax.grid(axis='y', alpha=0.3)
ax.legend(title='beta', fontsize=10)
plt.tight_layout()
plt.show()
```
