```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for better looking plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Comparison of Data Loading Methods: Avg Client Results', fontsize=16, fontweight='bold')

# Beta = 10 data
beta_10 = pubmed_gat[pubmed_gat['Beta'] == 10]

# Beta = 10000 data  
beta_10000 = pubmed_gat[pubmed_gat['Beta'] == 10000]

# Plot 1: Beta = 10, grouped by PE
ax1 = axes[0, 0]
data_loading_methods = ['zero_hop', 'adjacency', 'diffusion', 'full']
pe_true_10 = [beta_10[(beta_10['Data Loading'] == method) & (beta_10['pe'] == True)]['Avg Client Result'].iloc[0] 
              if len(beta_10[(beta_10['Data Loading'] == method) & (beta_10['pe'] == True)]) > 0 else 0 
              for method in data_loading_methods]
pe_false_10 = [beta_10[(beta_10['Data Loading'] == method) & (beta_10['pe'] == False)]['Avg Client Result'].iloc[0] 
               if len(beta_10[(beta_10['Data Loading'] == method) & (beta_10['pe'] == False)]) > 0 else 0 
               for method in data_loading_methods]

x = np.arange(len(data_loading_methods))
width = 0.35

bars1 = ax1.bar(x - width/2, pe_true_10, width, label='With PE', alpha=0.8)
bars2 = ax1.bar(x + width/2, pe_false_10, width, label='Without PE', alpha=0.8)

ax1.set_xlabel('Data Loading Method')
ax1.set_ylabel('Avg Client Result')
ax1.set_title('Beta = 10')
ax1.set_xticks(x)
ax1.set_xticklabels(data_loading_methods, rotation=45)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)
for bar in bars2:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)

# Plot 2: Beta = 10000, grouped by PE
ax2 = axes[0, 1]
pe_true_10000 = [beta_10000[(beta_10000['Data Loading'] == method) & (beta_10000['pe'] == True)]['Avg Client Result'].iloc[0] 
                 if len(beta_10000[(beta_10000['Data Loading'] == method) & (beta_10000['pe'] == True)]) > 0 else 0 
                 for method in data_loading_methods]
pe_false_10000 = [beta_10000[(beta_10000['Data Loading'] == method) & (beta_10000['pe'] == False)]['Avg Client Result'].iloc[0] 
                  if len(beta_10000[(beta_10000['Data Loading'] == method) & (beta_10000['pe'] == False)]) > 0 else 0 
                  for method in data_loading_methods]

bars3 = ax2.bar(x - width/2, pe_true_10000, width, label='With PE', alpha=0.8)
bars4 = ax2.bar(x + width/2, pe_false_10000, width, label='Without PE', alpha=0.8)

ax2.set_xlabel('Data Loading Method')
ax2.set_ylabel('Avg Client Result')
ax2.set_title('Beta = 10000')
ax2.set_xticks(x)
ax2.set_xticklabels(data_loading_methods, rotation=45)
ax2.legend()
ax2.grid(True, alpha=0.3)

# Add value labels on bars
for bar in bars3:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)
for bar in bars4:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)

# Plot 3: With PE comparison across Beta values
ax3 = axes[1, 0]
bars5 = ax3.bar(x - width/2, pe_true_10, width, label='Beta = 10', alpha=0.8)
bars6 = ax3.bar(x + width/2, pe_true_10000, width, label='Beta = 10000', alpha=0.8)

ax3.set_xlabel('Data Loading Method')
ax3.set_ylabel('Avg Client Result')
ax3.set_title('With Positional Encoding (PE = True)')
ax3.set_xticks(x)
ax3.set_xticklabels(data_loading_methods, rotation=45)
ax3.legend()
ax3.grid(True, alpha=0.3)

# Add value labels on bars
for bar in bars5:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)
for bar in bars6:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)

# Plot 4: Without PE comparison across Beta values
ax4 = axes[1, 1]
bars7 = ax4.bar(x - width/2, pe_false_10, width, label='Beta = 10', alpha=0.8)
bars8 = ax4.bar(x + width/2, pe_false_10000, width, label='Beta = 10000', alpha=0.8)

ax4.set_xlabel('Data Loading Method')
ax4.set_ylabel('Avg Client Result')
ax4.set_title('Without Positional Encoding (PE = False)')
ax4.set_xticks(x)
ax4.set_xticklabels(data_loading_methods, rotation=45)
ax4.legend()
ax4.grid(True, alpha=0.3)

# Add value labels on bars
for bar in bars7:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)
for bar in bars8:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Create a simplified comparison focusing on overall performance
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Simplified Comparison: Average Performance by Data Loading Method', fontsize=14, fontweight='bold')

# Calculate average performance across all conditions for each method
method_performance = {}
for method in data_loading_methods:
    method_data = pubmed_gat[pubmed_gat['Data Loading'] == method]
    avg_performance = method_data['Avg Client Result'].mean()
    method_performance[method] = avg_performance

# Sort methods by performance
sorted_methods = sorted(method_performance.items(), key=lambda x: x[1], reverse=True)
methods_sorted = [item[0] for item in sorted_methods]
performance_sorted = [item[1] for item in sorted_methods]

# Plot 1: Overall average performance
ax1 = axes[0]
colors = plt.cm.Set3(np.arange(len(methods_sorted)))
bars = ax1.bar(methods_sorted, performance_sorted, color=colors, alpha=0.8)
ax1.set_xlabel('Data Loading Method')
ax1.set_ylabel('Average Client Result')
ax1.set_title('Overall Average Performance')
ax1.grid(True, alpha=0.3)

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.005,
             f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 2: Performance by PE status
ax2 = axes[1]
pe_true_avg = []
pe_false_avg = []

for method in data_loading_methods:
    pe_true_data = pubmed_gat[(pubmed_gat['Data Loading'] == method) & (pubmed_gat['pe'] == True)]
    pe_false_data = pubmed_gat[(pubmed_gat['Data Loading'] == method) & (pubmed_gat['pe'] == False)]
    
    pe_true_avg.append(pe_true_data['Avg Client Result'].mean() if len(pe_true_data) > 0 else 0)
    pe_false_avg.append(pe_false_data['Avg Client Result'].mean() if len(pe_false_data) > 0 else 0)

x = np.arange(len(data_loading_methods))
width = 0.35

bars1 = ax2.bar(x - width/2, pe_true_avg, width, label='With PE', alpha=0.8, color='skyblue')
bars2 = ax2.bar(x + width/2, pe_false_avg, width, label='Without PE', alpha=0.8, color='lightcoral')

ax2.set_xlabel('Data Loading Method')
ax2.set_ylabel('Average Client Result')
ax2.set_title('Average Performance by PE Status')
ax2.set_xticks(x)
ax2.set_xticklabels(data_loading_methods, rotation=45)
ax2.legend()
ax2.grid(True, alpha=0.3)

# Add value labels
for bar in bars1:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.005,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.005,
             f'{height:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Print ranking summary
print("=== PERFORMANCE RANKING ===")
print("Overall Average Performance (highest to lowest):")
for i, (method, performance) in enumerate(sorted_methods, 1):
    print(f"{i}. {method.capitalize()}: {performance:.4f}")
    
print(f"\nBest performing method: {methods_sorted[0].capitalize()}")
print(f"Worst performing method: {methods_sorted[-1].capitalize()}")

# Check PE impact
print(f"\n=== POSITIONAL ENCODING IMPACT ===")
for method in data_loading_methods:
    pe_true_data = pubmed_gat[(pubmed_gat['Data Loading'] == method) & (pubmed_gat['pe'] == True)]
    pe_false_data = pubmed_gat[(pubmed_gat['Data Loading'] == method) & (pubmed_gat['pe'] == False)]
    
    if len(pe_true_data) > 0 and len(pe_false_data) > 0:
        pe_true_avg = pe_true_data['Avg Client Result'].mean()
        pe_false_avg = pe_false_data['Avg Client Result'].mean()
        improvement = pe_false_avg - pe_true_avg
        print(f"{method.capitalize()}: {'Better' if improvement > 0 else 'Worse'} without PE ({improvement:+.4f})")
```

```python
# Convert duration to seconds for better comparison
def duration_to_seconds(duration_str):
    """Convert HH:MM:SS format to total seconds"""
    try:
        parts = duration_str.split(':')
        hours = int(parts[0])
        minutes = int(parts[1]) 
        seconds = int(parts[2])
        return hours * 3600 + minutes * 60 + seconds
    except:
        return 0

# Add duration in seconds column
pubmed_gat['Duration_seconds'] = pubmed_gat['Duration'].apply(duration_to_seconds)

# Create efficiency comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Training Efficiency and Performance Analysis', fontsize=14, fontweight='bold')

# Plot 1: Average training duration by method
ax1 = axes[0]
avg_duration = []
for method in data_loading_methods:
    method_data = pubmed_gat[pubmed_gat['Data Loading'] == method]
    avg_dur = method_data['Duration_seconds'].mean() / 60  # Convert to minutes
    avg_duration.append(avg_dur)

bars = ax1.bar(data_loading_methods, avg_duration, color=plt.cm.viridis(np.linspace(0, 1, len(data_loading_methods))), alpha=0.8)
ax1.set_xlabel('Data Loading Method')
ax1.set_ylabel('Average Duration (minutes)')
ax1.set_title('Training Time Comparison')
ax1.grid(True, alpha=0.3)

# Add value labels
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{height:.1f}m', ha='center', va='bottom', fontweight='bold')

# Plot 2: Performance vs Duration scatter plot
ax2 = axes[1]
colors = {'zero_hop': 'red', 'adjacency': 'blue', 'diffusion': 'green', 'full': 'orange'}
for method in data_loading_methods:
    method_data = pubmed_gat[pubmed_gat['Data Loading'] == method]
    durations = method_data['Duration_seconds'] / 60  # Convert to minutes
    performances = method_data['Avg Client Result']
    ax2.scatter(durations, performances, label=method.capitalize(), 
               color=colors[method], alpha=0.7, s=100)

ax2.set_xlabel('Training Duration (minutes)')
ax2.set_ylabel('Avg Client Result')
ax2.set_title('Performance vs Training Time')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Efficiency score (Performance / Time)
ax3 = axes[2]
efficiency_scores = []
for method in data_loading_methods:
    method_data = pubmed_gat[pubmed_gat['Data Loading'] == method]
    avg_performance = method_data['Avg Client Result'].mean()
    avg_duration_hours = method_data['Duration_seconds'].mean() / 3600  # Convert to hours
    efficiency = avg_performance / avg_duration_hours  # Performance per hour
    efficiency_scores.append(efficiency)

# Sort by efficiency
method_efficiency = list(zip(data_loading_methods, efficiency_scores))
method_efficiency.sort(key=lambda x: x[1], reverse=True)
methods_eff_sorted = [item[0] for item in method_efficiency]
efficiency_sorted = [item[1] for item in method_efficiency]

bars = ax3.bar(methods_eff_sorted, efficiency_sorted, 
               color=plt.cm.plasma(np.linspace(0, 1, len(methods_eff_sorted))), alpha=0.8)
ax3.set_xlabel('Data Loading Method')
ax3.set_ylabel('Efficiency Score (Performance/Hour)')
ax3.set_title('Efficiency Ranking')
ax3.grid(True, alpha=0.3)

# Add value labels
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{height:.2f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Print efficiency analysis
print("=== EFFICIENCY ANALYSIS ===")
print("Efficiency Ranking (Performance per Hour):")
for i, (method, efficiency) in enumerate(method_efficiency, 1):
    print(f"{i}. {method.capitalize()}: {efficiency:.2f}")

print("\n=== TRAINING TIME SUMMARY ===")
for method in data_loading_methods:
    method_data = pubmed_gat[pubmed_gat['Data Loading'] == method]
    avg_duration_min = method_data['Duration_seconds'].mean() / 60
    min_duration_min = method_data['Duration_seconds'].min() / 60
    max_duration_min = method_data['Duration_seconds'].max() / 60
    print(f"{method.capitalize()}: Avg={avg_duration_min:.1f}min, Range={min_duration_min:.1f}-{max_duration_min:.1f}min")
```
