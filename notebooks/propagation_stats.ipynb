```python
# Define your directory paths
propagation_dirs = {
    'pubmed_beta_10_with_pe2': '/home/brian_bosho/FP/FP/federated-gnn/results/pubmed_gcn_final_beta_10_with_pe2/propagation_stats',
    'pubmed_beta_10_nope': '/home/brian_bosho/FP/FP/federated-gnn/results/pubmed_gcn_final_beta_10_nope/propagation_stats',
    'citeseer_ablation_pe': '/home/brian_bosho/FP/FP/federated-gnn/results/citeseer_ablation_final_pe/propagation_stats',
    'citeseer_ablation_no_pe': '/home/brian_bosho/FP/FP/federated-gnn/results/citeseer_ablation_final_no_pe/propagation_stats'
}

# Load data from all directories
client_duration_data = {}
for experiment_name, directory_path in propagation_dirs.items():
    try:
        df = get_client_durations_dataframe(directory_path)
        client_duration_data[experiment_name] = df
        print(f"✓ Loaded {experiment_name}: {df.shape[0]} files, {df.shape[1]} clients")
    except Exception as e:
        print(f"✗ Failed to load {experiment_name}: {e}")

print(f"\nTotal experiments loaded: {len(client_duration_data)}")
```

```python
# Analyze a specific experiment in detail
experiment_to_analyze = 'citeseer_ablation_pe'  # Change this as needed

if experiment_to_analyze in client_duration_data:
    df = client_duration_data[experiment_to_analyze]
    
    print(f"=== Analysis for {experiment_to_analyze} ===")
    print(f"Shape: {df.shape}")
    print(f"\nSummary Statistics:")
    print(df.describe())
    
    print(f"\nTotal propagation time per experiment run:")
    df['total_runtime'] = df.sum(axis=1)
    print(df['total_runtime'].describe())
    
    print(f"\nAverage runtime per client across all runs:")
    client_averages = df.mean()
    print(client_averages)
```
