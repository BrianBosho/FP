```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime

def compare_pe_performance(df, dataset_name, beta_value):
    """
    Creates bar plots comparing PE vs non-PE performance for different data loading options
    
    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the experiment results
    dataset_name : str
        Name of the dataset to filter by (e.g., 'Cora', 'Citeseer', 'Pubmed')
    beta_value : int
        Beta value to filter by (e.g., 1 or 10000)
    """
    # Filter data for the specified dataset and beta
    filtered_df = df[(df['dataset'] == dataset_name) & (df['beta'] == beta_value)]
    
    # Convert duration strings to minutes
    def duration_to_minutes(duration_str):
        time_obj = datetime.strptime(duration_str, '%H:%M:%S')
        return time_obj.hour * 60 + time_obj.minute + time_obj.second / 60
    
    filtered_df['duration_minutes'] = filtered_df['duration'].apply(duration_to_minutes)
    
    # Prepare data for plotting
    plot_data = filtered_df.pivot_table(
        index='data_loading_option', 
        columns='pe',
        values=['average_global_result', 'duration_minutes'],
        aggfunc='mean'
    ).reset_index()
    
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Get unique data loading options
    data_loading_options = filtered_df['data_loading_option'].unique()
    x = np.arange(len(data_loading_options))
    width = 0.35
    
    # Prepare data for barplots
    accuracy_no_pe = []
    accuracy_pe = []
    runtime_no_pe = []
    runtime_pe = []
    
    for option in data_loading_options:
        option_data = filtered_df[filtered_df['data_loading_option'] == option]
        
        # Get accuracy values
        no_pe_acc = option_data[option_data['pe'] == False]['average_client_result'].mean()
        pe_acc = option_data[option_data['pe'] == True]['average_client_result'].mean()
        accuracy_no_pe.append(no_pe_acc)
        accuracy_pe.append(pe_acc)
        
        # Get runtime values
        no_pe_runtime = option_data[option_data['pe'] == False]['duration_minutes'].mean()
        pe_runtime = option_data[option_data['pe'] == True]['duration_minutes'].mean()
        runtime_no_pe.append(no_pe_runtime)
        runtime_pe.append(pe_runtime)
    
    # Plot accuracy comparison
    ax1.bar(x - width/2, accuracy_no_pe, width, label='No PE', color='tab:blue')
    ax1.bar(x + width/2, accuracy_pe, width, label='With PE', color='tab:orange')
    ax1.set_ylabel('Accuracy')
    ax1.set_title(f'Accuracy vs. PE ({dataset_name}, β={beta_value})')
    ax1.set_xticks(x)
    ax1.set_xticklabels([opt.replace('_', '-') for opt in data_loading_options])
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)
    
    # Set y-axis limits to highlight differences
    y_min = max(0.5, min(min(accuracy_no_pe), min(accuracy_pe)) - 0.05)
    y_max = min(1.0, max(max(accuracy_no_pe), max(accuracy_pe)) + 0.05)
    ax1.set_ylim(y_min, y_max)
    
    # Plot runtime comparison
    ax2.bar(x - width/2, runtime_no_pe, width, label='No PE', color='tab:blue')
    ax2.bar(x + width/2, runtime_pe, width, label='With PE', color='tab:orange')
    ax2.set_ylabel('Minutes')
    ax2.set_title(f'Runtime vs. PE ({dataset_name}, β={beta_value})')
    ax2.set_xticks(x)
    ax2.set_xticklabels([opt.replace('_', '-') for opt in data_loading_options])
    ax2.legend()
    ax2.grid(True, linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    
    return fig

# Example usage:
# fig = compare_pe_performance(df, 'Cora', 1)
# plt.show()
```

```python
def compare_pe_by_beta(df, dataset_name):
    """
    Creates side-by-side bar plots comparing PE vs no PE accuracy for different data loading options,
    with separate plots for IID (beta=1) and non-IID (beta=10000) settings.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the experiment results
    dataset_name : str
        Name of the dataset to filter by (e.g., 'Cora', 'Citeseer', 'Pubmed')
    """
    # Filter data for the specified dataset
    filtered_df = df[df['dataset'] == dataset_name]
    
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # Get unique data loading options
    data_loading_options = filtered_df['data_loading_option'].unique()
    x = np.arange(len(data_loading_options))
    width = 0.35
    
    # Plot for beta=1 (IID)
    beta1_no_pe = []
    beta1_pe = []
    for option in data_loading_options:
        no_pe_acc = filtered_df[(filtered_df['data_loading_option'] == option) & 
                               (filtered_df['pe'] == False) & 
                               (filtered_df['beta'] == 1)]['average_client_result'].mean()
        pe_acc = filtered_df[(filtered_df['data_loading_option'] == option) & 
                           (filtered_df['pe'] == True) & 
                           (filtered_df['beta'] == 1)]['average_client_result'].mean()
        beta1_no_pe.append(no_pe_acc)
        beta1_pe.append(pe_acc)
    
    # Plot for beta=10000 (non-IID)
    beta10000_no_pe = []
    beta10000_pe = []
    for option in data_loading_options:
        no_pe_acc = filtered_df[(filtered_df['data_loading_option'] == option) & 
                               (filtered_df['pe'] == False) & 
                               (filtered_df['beta'] == 10000)]['average_client_result'].mean()
        pe_acc = filtered_df[(filtered_df['data_loading_option'] == option) & 
                           (filtered_df['pe'] == True) & 
                           (filtered_df['beta'] == 10000)]['average_client_result'].mean()
        beta10000_no_pe.append(no_pe_acc)
        beta10000_pe.append(pe_acc)
    
    # Plot beta=1 chart
    ax1.bar(x - width/2, beta1_no_pe, width, label='No PE', color='tab:blue')
    ax1.bar(x + width/2, beta1_pe, width, label='With PE', color='tab:orange')
    ax1.set_ylabel('Client Accuracy')
    ax1.set_title(f'Accuracy vs. PE ({dataset_name}, β=1, IID)')
    ax1.set_xticks(x)
    ax1.set_xticklabels([opt.replace('_', '-') for opt in data_loading_options])
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)
    
    # Get y-axis limits that work for both plots
    all_values = beta1_no_pe + beta1_pe + beta10000_no_pe + beta10000_pe
    ymin = max(0.50, min(all_values) - 0.02)
    ymax = min(1.0, max(all_values) + 0.02)
    
    ax1.set_ylim(ymin, ymax)
    
    # Plot beta=10000 chart
    ax2.bar(x - width/2, beta10000_no_pe, width, label='No PE', color='tab:blue')
    ax2.bar(x + width/2, beta10000_pe, width, label='With PE', color='tab:orange')
    ax2.set_ylabel('Client Accuracy')
    ax2.set_title(f'Accuracy vs. PE ({dataset_name}, β=10000, non-IID)')
    ax2.set_xticks(x)
    ax2.set_xticklabels([opt.replace('_', '-') for opt in data_loading_options])
    ax2.legend()
    ax2.grid(True, linestyle='--', alpha=0.7)
    ax2.set_ylim(ymin, ymax)
    
    plt.tight_layout()
    
    return fig

# Example usage:
# fig = compare_pe_by_beta(df, 'Cora')
# plt.show()
```
