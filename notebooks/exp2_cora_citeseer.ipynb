```python
import os
import pathlib

def find_repo_root(start=None):
    p = pathlib.Path(start or os.getcwd()).resolve()
    for parent in [p, *p.parents]:
        if (parent / "conf" / "base.yaml").exists():
            return parent
    return p

PROJECT_ROOT = find_repo_root()
path = str(PROJECT_ROOT / "results" / "results_summary" / "cora_baseline")
```

```python
# Merge all DataFrames into one comprehensive DataFrame
import pandas as pd

# First, add identifier columns to each DataFrame to distinguish configurations
dataframes_to_merge = []

# Cora experiments
if 'df_cora_gcn_beta10000_no_pe_nzh' in locals():
    df_temp = df_cora_gcn_beta10000_no_pe_nzh.copy()
    df_temp['dataset'] = 'cora'
    df_temp['beta'] = 10000
    df_temp['positional_encoding'] = False
    df_temp['hop_type'] = 'non_zero_hop'
    dataframes_to_merge.append(df_temp)

if 'df_cora_gcn_beta1_no_pe_nzh' in locals():
    df_temp = df_cora_gcn_beta1_no_pe_nzh.copy()
    df_temp['dataset'] = 'cora'
    df_temp['beta'] = 1
    df_temp['positional_encoding'] = False
    df_temp['hop_type'] = 'non_zero_hop'
    dataframes_to_merge.append(df_temp)

if 'df_cora_gcn_pe_beta10000_nzh' in locals():
    df_temp = df_cora_gcn_pe_beta10000_nzh.copy()
    df_temp['dataset'] = 'cora'
    df_temp['beta'] = 10000
    df_temp['positional_encoding'] = True
    df_temp['hop_type'] = 'non_zero_hop'
    dataframes_to_merge.append(df_temp)

if 'df_cora_gcn_pe_beta1_nzh' in locals():
    df_temp = df_cora_gcn_pe_beta1_nzh.copy()
    df_temp['dataset'] = 'cora'
    df_temp['beta'] = 1
    df_temp['positional_encoding'] = True
    df_temp['hop_type'] = 'non_zero_hop'
    dataframes_to_merge.append(df_temp)

if 'cora_gcn_zh_no_pe' in locals():
    df_temp = cora_gcn_zh_no_pe.copy()
    df_temp['dataset'] = 'cora'
    df_temp['beta'] = None  # Will need to infer from other columns if needed
    df_temp['positional_encoding'] = False
    df_temp['hop_type'] = 'zero_hop'
    dataframes_to_merge.append(df_temp)

if 'cora_gcn_zh_pe' in locals():
    df_temp = cora_gcn_zh_pe.copy()
    df_temp['dataset'] = 'cora'
    df_temp['beta'] = None  # Will need to infer from other columns if needed
    df_temp['positional_encoding'] = True
    df_temp['hop_type'] = 'zero_hop'
    dataframes_to_merge.append(df_temp)

# Citeseer experiments
if 'df_citeseer_gcn_no_pe' in locals():
    df_temp = df_citeseer_gcn_no_pe.copy()
    df_temp['dataset'] = 'citeseer'
    df_temp['beta'] = None  # Will need to infer from other columns if needed
    df_temp['positional_encoding'] = False
    df_temp['hop_type'] = 'mixed'  # Based on the output, seems to have multiple hop types
    dataframes_to_merge.append(df_temp)

if 'df_citeseer_gcn_pe' in locals():
    df_temp = df_citeseer_gcn_pe.copy()
    df_temp['dataset'] = 'citeseer'
    df_temp['beta'] = None  # Will need to infer from other columns if needed
    df_temp['positional_encoding'] = True
    df_temp['hop_type'] = 'mixed'  # Based on the output, seems to have multiple hop types
    dataframes_to_merge.append(df_temp)

# Merge all DataFrames
if dataframes_to_merge:
    df_all_experiments = pd.concat(dataframes_to_merge, ignore_index=True)
    print(f"Successfully merged {len(dataframes_to_merge)} DataFrames into one with {len(df_all_experiments)} total rows")
    print(f"Columns: {list(df_all_experiments.columns)}")
    print(f"Dataset distribution: {df_all_experiments['dataset'].value_counts()}")
    print(f"Positional encoding distribution: {df_all_experiments['positional_encoding'].value_counts()}")
else:
    print("No DataFrames found to merge")

df_all_experiments
```
